```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))


content = """15 w1  
h1  
40 w5  
o1  
20 w2  
45 w6  
0.01  
25 w3  
50 w7  
30 w4  
h2  
55 w8  
o2  
.99  
b1.35  
b2.60  
"""


weights = {}
biases = {}
targets = []

for line in content.split('\n'):
    line = line.strip()
    if not line:
        continue
    parts = line.split()
    if parts[0].replace('.', '', 1).isdigit():
        value = float(parts[0])
        if len(parts) > 1 and parts[1].startswith('w'):
            weights[parts[1]] = value
        else:
            targets.append(value)
    elif parts[0].startswith('b'):
        bias_value = float(parts[0][1:])
        biases[parts[0]] = bias_value

W1 = np.array([
    [weights['w1'], weights['w3']],
    [weights['w2'], weights['w4']]
])


W2 = np.array([
    [weights['w5'], weights['w7']],
    [weights['w6'], weights['w8']]
])


b1 = np.array([biases['b1.35'], biases['b2.60']])  # Hidden layer biases
b2 = np.zeros(2)  # Output layer biases (assumed zero)

x = np.array([1, 1])
y = np.array(targets)

z1 = np.dot(x, W1) + b1
a1 = sigmoid(z1)
z2 = np.dot(a1, W2) + b2
a2 = sigmoid(z2)


delta2 = (a2 - y) * sigmoid_derivative(z2)
dW2 = np.outer(a1, delta2)
db2 = delta2

delta1 = np.dot(delta2, W2.T) * sigmoid_derivative(z1)
dW1 = np.outer(x, delta1)
db1 = delta1

print("Gradients for W1:")
print(dW1)
print("\nGradients for b1:")
print(db1)
print("\nGradients for W2:")
print(dW2)
print("\nGradients for b2:")
print(db2)
```
